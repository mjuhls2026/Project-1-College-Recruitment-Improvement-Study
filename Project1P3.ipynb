```python
#Import pandas and MatPlotLib into the notebook
import pandas as pd
import matplotlib.pyplot as plt
```

```python
df = pd.read_csv("C:/Users/juhls.matthew/Documents/Data Study Project/Cleaned_Majors.csv")
```

```python
df.info()
```

```python
# Bar graph for how students found out about CCM
# I looked to the textbook and powerpoint before turning to Ai because i wanted to see if there was any helpful examples,
# I used Ai to figure out how to only accept the 'Yes' answers for each column and only do the graph based on the 'Hear' columns.
# I also used Ai to figure out how to make the bottom text not smushed and readable, because before it was all squished together,
# so it gave me xticks, which i did notice in was in the powerpoint, i just didnt know what kind of info went into the parenthesis.
hear = [col for col in df.columns if col.startswith("Hear")]

yes_counts = []
for col in hear:
    count = (df[col] == "Yes").sum()
    yes_counts.append(count)
    
plt.bar(hear, yes_counts)
plt.title("Most Effective Ways Students Heard about CCM", fontsize=18)
plt.xlabel("Information Outreach", fontsize = 12)
plt.ylabel("Number of Answers", fontsize = 12)
plt.xticks(rotation=90)
plt.show()
```

```python
# Pie chart to show ratings of the courses students took.
# The number next to the section represents the rating that the students gave 1-5.
# Same deal as the graph above, I needed to figure out how to put the ratings into the pie chart
rating = df["Rating (1-5)"].value_counts()

plt.pie(rating, labels=rating.index)
plt.title("Course Ratings (1-5)", fontsize = 18)
plt.show()
```
